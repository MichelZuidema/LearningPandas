<h1>Combining</h1>

```python
can_yt = pd.read_csv("datasets/CAvideos.csv")
gb_yt = pd.read_csv("datasets/GBvideos.csv")

pd.concat([can_yt, gb_yt])
```

```python
# Get videos that were trending on the same day
left = can_yt.set_index(['title'], 'trending_date')
right = gb_yt.set_index(['title'], 'trending_date')

left.join(right, lsuffix='_CAN', rsuffix='_UK')
```
